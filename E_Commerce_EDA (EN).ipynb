```python
# Outlier Detection
for col in ['price','total_amount', 'shipping_cost']:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
summary = {}

for col in ['price','total_amount', 'shipping_cost']:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower) | (df[col] > upper)]

    summary[col] = {
        'Q1': Q1, 'Q3': Q3, 'IQR': IQR,
        'Lower bound': lower,
        'Upper bound': upper,
        'Num outliers': len(outliers),
        'Pct outliers': len(outliers)/len(df)*100,
        'Max value': df[col].max()
    }

summary_df = pd.DataFrame(summary).T
print(summary_df)
```

##### **1. Boxplot Analysis**
A. Price (Product Price)
- The distribution is **strongly right-skewed**: most products are priced below 150, but the range extends up to 2,500.
- IQR = 114.26, upper bound = 302.34,meaning that any product priced above 302 is considered an outlier, accounting for approximately **~10.5%** of the data.
- The outliers are densely concentrated between 300–1,500, forming a long right tail → **indicating the presence of a premium product segment**  
  
B. Total Amount (Order Value)
- The distribution is clearly right-skewed, ranging from 0 to over 12,000.
- IQR = 148.82, upper bound = 391076
- With approximately 11%of data points classified as outliers, max = 12,931.8  
→ These outliers are mainly associated with **high-value transactions**, likely resulting from customers purchasing expensive or multiple items. **Hence, they are not anomalies, but rather valid observations representing premium or bulk orders.**
  
C. Shipping Cost
- The distribution is nearly symmetric, with a median around 6, IQR = 3.4, upper bound ~ 12.95
- The few values above this threshold are reasonable, as **larger orders typically incur higher shipping costs.**
- Only **0.14%** of observations are outliers, max = 15.65, which is negligible and justifiable (due to long-distance or large-order deliveries).

##### **2. Outlier Evaluation Criteria**
- Outlier ratio > 10% -> likely represents a distinct market segment; should not be removed.
- Outlier ratio < 2% -> likely noise or errors; can be removed.  
  
Based on the analysis, the outlier ratios for **total_amount (11%)** and **price (10.5%)** suggest the presence of a higher-value product segment, possibly premium goods. Meanwhile, **shipping_cost** has a very low outlier rate (**0.14%**) indicating no significant distortion in the data.  
→ We hypothesize that these high-value outliers may correspond to Electronic products, and we will verify whether the Electronics category is indeed responsible for these observations in the next step.

```python
df['is_outlier_amount'] = (df['total_amount'] > summary['total_amount']['Upper bound'])
df['is_outlier_price'] = (df['price'] > summary['price']['Upper bound'])

pd.crosstab(df['category'], df['is_outlier_amount'], normalize='index') * 100
```

```python
pd.crosstab(df['category'], df['is_outlier_price'], normalize='index') * 100
```

**Conclusion:**  
From the table, we observe that Electronics accounts for approximately **~43-46%** of the outliers in both **total_amount** and **price.**  
→ This confirms that these are **contextual outliers**, not data entry errors. **They represent genuine high-value transactions, and therefore should not be removed from the dataset.**

**Insight:**
Grocery losses are not discount-driven but stem from the operational model and inherently thin industry margins.
